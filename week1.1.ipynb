# Split Skill Column by Import re

```python
skill1=df.Skill
```

```python
skill1=skill1.tolist() #Converting Values to List and declaring in Skill1
```

```python
skill1
```

# Extract the Skill values between the single quote

```python
import re

patt = r"'(.*?)'"  # Matches any text between single quotes.
matc=[]            # this empty list used to store the skill values

for i in range(len(skill1)):
    temp = re.findall(patt, skill1[i])
    matc.extend(temp)
```

```python
print(len(matc))   # we have 15469 number of skills
print(matc)
```

# Remove repeated values in this list

```python
print(type(matc))         
print(len(set(matc)))  # we have 300 unique values 

matc=set(matc)         # type casting list to set for unique values of skills
print(type(matc))

matc=list(matc)        # again turn the set into list for performing operations by index number of values
print(type(matc))

matc.sort()            # sort the list into ascending order
print(matc)
```

# Here we Split the skills into columns and fill the value with 1 if present and 0 if not

#Create a dictionary to store the results for each skill


```python
skill_dict = {}
for skill in matc:
    skill_dict[skill] = []
```

#Iterate over each row in the DataFrame and extract the skills from the 'skills' column


```python
for i, row in df.iterrows():
    skills = eval(row['Skill'])
    for skill in matc:
        if skill.lower() in [s.lower() for s in skills]:
            skill_dict[skill].append(1)
        else:
            skill_dict[skill].append(0)
```

#Create a new DataFrame from the skill_dict dictionary


```python
skill_df = pd.DataFrame(skill_dict)
```

#Concatenate the original DataFrame and the skill_df DataFrame


```python
df = pd.concat([df, skill_df], axis=1)
```

```python
df = df.drop(columns=['Skill'])  # drop Skill column  
df.head(5)
```
